```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv("/Users/ravitiwari/Desktop/MY Data/work/garments_worker_productivity.csv",index_col="date", parse_dates=True)
Data
```

```python
Data['department'] = Data['department'].map({ 'finishing ':'finishing','sweing':'sweing','finishing':'finishing'})
```

```python
Data.isnull().sum()
```

```python
Data.shape
```

```python
Data.columns
```

```python
plt.figure(figsize = (20,6))
ax=sns.lineplot(y='targeted_productivity',x='date' ,data =Data,)
ax=sns.lineplot(y= 'actual_productivity',x='date',data=Data)
ax.set(ylabel = 'Productivity')
plt.show()
```

```python
plt.figure(figsize = (6,6))
ax=sns.violinplot(y='targeted_productivity',x='department' ,data =Data, box = True, Point = 'all')
ax=sns.violinplot(y= 'actual_productivity',x='department',data=Data,box = True)
ax.set(ylabel = 'Productivity')
plt.show()
```

```python
plt.figure(figsize = (20,10))
ax=sns.lineplot(y='targeted_productivity',x='date' ,data =Data,hue= 'department' )
ax.set(ylabel = 'Target Productivity')
plt.show()
```

```python
plt.figure(figsize = (20,10))
ax=sns.lineplot(y= 'actual_productivity',x='date',data=Data,hue= 'department')
ax.set(ylabel = 'actual Productivity')
plt.show()
```
